```python
import numpy as np
from glob import glob
import librosa
import librosa.display
import IPython.display as ipd
from IPython.display import Audio, display
from itertools import cycle
import pygame
import soundfile as sf
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
from scipy.fft import fft
```

```python
y, sr = librosa.load("AI_SPEECH.wav")
librosa.display.waveshow(y)
```

```python
Audio("AI_SPEECH.wav")
vowel1=y[25000:29000]
Audio(vowel1, rate = sr)
```

```python
fft_result = fft(vowel1)

# Calculate the frequency bins
n = len(fft_result)
frequencies = np.fft.fftfreq(n, d=1/sr)
positive_frequencies = frequencies[:n//2]
magnitude_spectrum = np.abs(fft_result[:n//2])

# Plot the amplitude spectrum
plt.figure(figsize=(10, 4))
plt.plot(positive_frequencies, magnitude_spectrum)
plt.title('Amplitude Spectrum of the Signal Snippet')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()

fft_result = fft(vowel2)

# Calculate the frequency bins
n = len(fft_result)
frequencies = np.fft.fftfreq(n, d=1/sr)
positive_frequencies = frequencies[:n//2]
magnitude_spectrum = np.abs(fft_result[:n//2])

# Plot the amplitude spectrum
plt.figure(figsize=(10, 4))
plt.plot(positive_frequencies, magnitude_spectrum)
plt.title('Amplitude Spectrum of the Signal Snippet')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()

# Repeat for a few vowel sounds
vowel_snippets = [
    (25000, 29000),(31000, 35000)
]

for start, end in vowel_snippets:
    vowel_snippet = y[start:end]
    display(Audio(vowel_snippet, rate=sr))

    # Perform FFT on the vowel snippet
    fft_result_vowel = fft(vowel_snippet)

    # Calculate the frequency bins
    n_vowel = len(fft_result_vowel)
    frequencies_vowel = np.fft.fftfreq(n_vowel, d=1/sr)
    positive_frequencies_vowel = frequencies_vowel[:n_vowel//2]
    magnitude_spectrum_vowel = np.abs(fft_result_vowel[:n_vowel//2])

    # Plot the amplitude spectrum for the vowel snippet
    plt.figure(figsize=(10, 4))
    plt.plot(positive_frequencies_vowel, magnitude_spectrum_vowel)
    plt.title('Amplitude Spectrum of the Vowel Sound')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)
    plt.show()
```
