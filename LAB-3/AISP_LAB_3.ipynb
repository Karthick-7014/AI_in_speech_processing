```python
import librosa as lb
import soundfile as sf
import IPython.display as ipd

y, sr = lb.load("AI_SPEECH.wav", sr=None)

y_trim, _ = lb.effects.trim(y)
t_audio_path = "trimmed_speech.wav"
sf.write(t_audio_path, y_trim, sr)

trimmed_audio, _ = lb.load(t_audio_path, sr=None)

print("Original Audio:")
ipd.display(ipd.Audio(y, rate=sr))

print("Trimmed Audio:")
ipd.display(ipd.Audio(trimmed_audio, rate=sr))
```

```python
split_intervals = lb.effects.split(y, top_db=30)

for i, (start, end) in enumerate(split_intervals):
    split_audio = y[start:end]
    split_audio_p = f"split_{i}.wav"
    sf.write(split_audio_p, split_audio, sr)
    split_audio, _ = lb.load(split_audio_p, sr=None)
    print(f"Split Audio {i}:")
    ipd.display(ipd.Audio(split_audio, rate=sr))
```
